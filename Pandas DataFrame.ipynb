## Selecting rows from a DataFrame

- **loc** - searches using index labels

```python
# single row
movies.iloc[0]
```

```python
type(movies.iloc[0])
```

```python
ipl.iloc[4]
```

```python
# multiple row
movies.iloc[0:5]
```

```python
ipl.iloc[:5]
```

```python
type(ipl.iloc[:5])
```

```python
ipl.iloc[0:20:2]
```

```python
# fancy indexing
movies.iloc[[2,8,9,15]]
```

```python
# using Dict
 
student_dict = {
    'name' :['rahul','ankit','rupesh','amitabh','ramesh','ankita','rishabh'],
    'sub_1' :[74,58,45,74,0,0,74],
    'sub_2' :[80,70,45,85,0,0,80],
    'package':[4,6,4,5,0,0,5]    
}

students = pd.DataFrame(student_dict)
students.set_index('name', inplace=True)
students
```

```python
# loc
students.loc['ramesh']
```

```python
students.iloc[1:5]
```

```python
students.loc['ankit':'ankita']
```

```python
students.loc[['ankit','ankita','amitabh']]
```

```python
students.iloc[[1,6,2]]
```
